### Schedule Execution


**Task:** Schedule the multi-task job (Bronze → Silver → Gold)

The workflow created using Databricks Jobs can be scheduled to run automatically
at a defined time or frequency (for example: daily, hourly, or weekly).

In Databricks Workflows, scheduling enables:
- Automated execution without manual intervention
- Consistent pipeline runs at fixed intervals
- Reliable orchestration of dependent tasks (Bronze → Silver → Gold)

For this workflow, the schedule configuration is managed through the
Databricks Jobs interface, where the job can be triggered manually or
configured with a time-based schedule as per requirements.

This completes the end-to-end orchestration lifecycle:
**parameterization → multi-task job → dependencies → scheduling**.
